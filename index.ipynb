```python
coffee_sales = df.groupby("coffee_name")["money"].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
coffee_sales.plot(kind="bar", color="skyblue")
plt.title("Total Sales by Coffee Type")
plt.ylabel("Total Money")
plt.xlabel("Coffee Name")
plt.xticks(rotation=45)
plt.show()
```

This plot shows the total sales by coffee type.Here,this plot shows that Americano with Milk, Latte, and Cappuccino bring the highest sales, while Espresso and Cocoa sell the least.

```python
plt.figure(figsize=(8,5))
sns.barplot(x='coffee_name', y='money', data=df, estimator=sum, order=df.groupby('coffee_name')['money'].sum().sort_values(ascending=False).index)
plt.title("Total Revenue by Coffee Type")
plt.xticks(rotation=45)
plt.ylabel("Total Revenue")
plt.show()
```

This plot shows the total revenue earned from each coffee type. Here,Americano and latte which are milk based coffees accelerate more revenue than off espresso and cocoa which contributes the least revenue.

```python
daily_sales = df.groupby("date")["money"].sum()
plt.figure(figsize=(10,5))
daily_sales.plot(marker="o", linestyle="-", color="green")
plt.title("Daily Sales Trend")
plt.ylabel("Total Money")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```

This plot shows the daily sales trend,ie,total money earned per day.
Here,this plot shows that sales fluctutates daily,mostly between 100-400 and also going very high around 600 which brings higher revenue than normal.

```python
top_customers = df.groupby("card")["money"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top_customers.plot(kind="bar", color="purple")
plt.title("Top 5 Customers by Spend")
plt.xlabel("Customer ID")
plt.ylabel("Total Spent")
plt.show()
```

This plot shows the top five customers ranked by the total amount of money they have spent.the customer with the ID ANON-0000-0000-0012 is the top spender,he spend over $2500.this chart shows the highest spenders to the fifth highest spender from left to right in a descending manner

```python
plt.figure(figsize=(8,4))
df.groupby(df['datetime'].dt.hour)['money'].sum().plot(kind='line', marker='o')
plt.title("Revenue Trend by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Total Revenue")
plt.show()
```

```python
df["day_of_week"] = df["datetime"].dt.day_name()
dow_sales = df.groupby("day_of_week")["money"].sum()
dow_sales = dow_sales.reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.figure(figsize=(10, 5))
dow_sales.plot(kind="bar", color="skyblue")
plt.title("Sales by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Total Sales")
plt.show()
```

```python
americano_sales = df[df['coffee_name'].str.contains("Americano", case=False, na=False)].loc[:,:]
```

```python
americano_sales['hour'] = americano_sales['datetime'].dt.hour
hourly_americano = americano_sales.groupby('hour')['money'].sum().reset_index()

plt.figure(figsize=(10,5))
plt.plot(hourly_americano['hour'], hourly_americano['money'], marker='o', color='brown')
plt.title("Hourly Sales Trend - Americano")
plt.xlabel("Hour of Day")
plt.ylabel("Total Americano Sales")
plt.xticks(range(0,24))
plt.grid(True)
plt.show()
```

```python

df['hour'] = df['datetime'].dt.hour

sales_by_item_hour = df.groupby(['coffee_name','hour'])['money'].sum().reset_index()


pivot = sales_by_item_hour.pivot(index='coffee_name', columns='hour', values='money').fillna(0)

plt.figure(figsize=(14,6))
sns.heatmap(pivot, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Coffee Sales by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Coffee Name")
plt.show()
```
